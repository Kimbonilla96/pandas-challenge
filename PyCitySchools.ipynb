```python
import pandas as pd
from pathlib import Path
```

# District Summary 

```python
# Load in file
# Store filepath in a variable
school_file = Path("PyCitySchools/schools_complete.csv")
```

```python
student_file = Path("PyCitySchools/students_complete.csv")
```

```python
# Read and display the CSV with Pandas
school_file_df = pd.read_csv(school_file)
```

```python
student_file_df = pd.read_csv(student_file)
```

```python
reorganized_df = pd.merge(school_file_df, student_file_df, how= "left", on=["school_name", "school_name"])
```

```python
reorganized_df.head()
```

```python
#organized the column names to make the DataFrame look cleaner and easier to interpret
df2 = reorganized_df[['Student ID',
                      'School ID',
                      'gender',
                      'student_name',
                      'school_name',
                      'grade',
                      'reading_score',
                      'math_score','type',
                      'size',
                      'budget'
        ]]
df2.head()
```

```python
#Total number of unique schools
unique_schools = df2['school_name'].nunique()
unique_schools
```

```python
#Total students
Total_students = df2['student_name'].count()
Total_students
```

```python
#total budget
total_budget =df2['budget'].unique().sum()
total_budget
```

```python
#Average math score
Average_math_score =round(df2['math_score'].mean(),2)
Average_math_score
```

```python
#Average reading score
Average_reading_score = round(df2['reading_score'].mean(),2)
Average_reading_score
```

```python
# % passing math (the percentage of students who passed math)
number_passing_math = df2.loc[df2['math_score'] >= 70]

percentage_passing_math = round((len(number_passing_math)/len(df2)) * 100, 2)
percentage_passing_math
```

```python
# % passing reading (the percentage of students who passed reading)
number_passing_reading = df2.loc[df2['reading_score'] >= 70]

percentage_passing_reading = round((len(number_passing_reading)/len(df2)) * 100, 2)
percentage_passing_reading
```

```python
# % overall passing (the percentage of students who passed math AND reading)
student_passing_both_subjects = df2.loc[(df2['math_score'] >= 70) & (df2['reading_score'] >= 70)]

percentage_passing_both = round((len(student_passing_both_subjects)/len(df2)) * 100, 2)

percentage_passing_both
```

```python
#district's key metrics in a DataFrame
district_summary = [{
    "Total number of unique schools": unique_schools,
    "Total Students": Total_students,
    "Total Budget": total_budget,
    "Average math score": Average_math_score,
    "Average reading score": Average_reading_score,
    "Percentage Passing math": percentage_passing_math,
    "Percentage Passing reading": percentage_passing_reading,
    "Percantage Overall Passing": percentage_passing_both
    
}]

pd.DataFrame(district_summary)
```

# School Summary

```python
#School type
school_type = df2.groupby('school_name')["type"].first()
school_type
```

```python
#Total students
total_student = df2['school_name'].value_counts()
total_student
```

```python
#Total school budget
total_school_budget = df2.groupby(['school_name'])['budget'].mean()

total_school_budget
```

```python
#Per student budget
per_student_budget = (total_school_budget/total_student)
per_student_budget
```

```python
#Average math score for each school
mean_math_scores = df2.groupby(['school_name'])['math_score'].mean()
mean_math_scores
```

```python
#Average reading score for each school
mean_reading_scores = df2.groupby(['school_name'])['reading_score'].mean()
mean_reading_scores
```

```python
# % passing math (the percentage of students who passed math)

# Filter the DataFrame for students passing math
passing_math_df = df2[df2['math_score'] >= 70]

# Group the passing math DataFrame by school name
school_passing_math = passing_math_df.groupby('school_name')

# Calculate the number of students passing math for each school
passing_math_count = school_passing_math['Student ID'].count()

# Calculate the percentage of students passing math for each school
percentage_passing_math = (passing_math_count / total_student) * 100

percentage_passing_math
```

```python
#% passing reading (the percentage of students who passed reading)

# Filter the DataFrame for students passing reading
passing_reading_df = df2[df2['reading_score'] >= 70]

# Group the passing reading DataFrame by school name
school_passing_reading = passing_reading_df.groupby('school_name')

# Calculate the number of students passing reading for each school
passing_reading_count = school_passing_reading['Student ID'].count()

# Calculate the percentage of students passing reading for each school
percentage_passing_reading = (passing_reading_count / total_student) * 100

percentage_passing_reading
```

```python
#% overall passing (the percentage of students who passed math AND reading)

# Filter the DataFrame for students passing reading and math
passing_both_df = df2[(df2['math_score'] >= 70) & (df2['reading_score'] >= 70)]

# Group the passing math and reading DataFrame by school name
school_passing_both = passing_both_df.groupby('school_name')

# Calculate the number of students passing math and reading for each school
passing_both_count = school_passing_both['Student ID'].count()

# Calculate the percentage of students passing math anf reading for each school
percentage_passing_both = (passing_both_count/ total_student) * 100

percentage_passing_both
```

```python
school_summary = pd.DataFrame({
    "School Type": school_type,
    "Total Students": total_student.values,
    "Total school budget": total_school_budget,
    "Per student budget": per_student_budget,
    "Average math score": mean_math_scores,
    "Average reading score": mean_reading_scores,
    "Percentage Passing math": percentage_passing_math,
    "Percentage Passing reading": percentage_passing_reading,
    "Percantage Overall Passing": percentage_passing_both
    
})

school_summary["Total school budget"] = school_summary["Total school budget"].map("${:,.2f}".format)
school_summary["Per student budget"] = school_summary["Per student budget"].map("${:,.2f}".format)

school_summary["Average math score"] = school_summary["Average math score"].map("{:,.2f}".format)
school_summary["Average reading score"] = school_summary["Average reading score"].map("{:,.2f}".format)

school_summary["Percentage Passing math"] = school_summary["Percentage Passing math"].map("{:,.2f}".format)
school_summary["Percentage Passing reading"] = school_summary["Percentage Passing reading"].map("{:,.2f}".format)
school_summary["Percantage Overall Passing"] = school_summary["Percantage Overall Passing"].map("{:,.2f}".format)

school_df = pd.DataFrame(school_summary)
school_df
```

# Scores by School Spending

```python
#Create a table that breaks down school performance based on average spending ranges (per student).
#Use the code provided below to create four bins with reasonable cutoff values to group school spending.

spending_bins = [0, 585, 630, 645, 680]
labels = ['<$585', '$585-630', '$630-645', '$645-680']
```

```python
# Create a copy of the school summary since it has the "Per Student Budget"
school_spending_df = school_df.copy()
school_spending_df
```

```python
school_spending_df["Spending Ranges(Per Student)"] = pd.cut(per_student_budget, bins=spending_bins, labels=labels, right = False, include_lowest = True)
school_spending_df
```

```python
school_spending_df["Average math score"] = pd.to_numeric(school_spending_df["Average math score"])

school_spending_df["Average reading score"] = pd.to_numeric(school_spending_df["Average reading score"])

school_spending_df["Percentage Passing math"] = pd.to_numeric(school_spending_df["Percentage Passing math"])

school_spending_df["Percentage Passing reading"] = pd.to_numeric(school_spending_df["Percentage Passing reading"])

school_spending_df["Percantage Overall Passing"] = pd.to_numeric(school_spending_df["Percantage Overall Passing"])
```

```python
spending_math_scores = school_spending_df.groupby(["Spending Ranges(Per Student)"])["Average math score"].mean()
spending_reading_scores = school_spending_df.groupby(["Spending Ranges(Per Student)"])["Average reading score"].mean()
spending_passing_math = school_spending_df.groupby(["Spending Ranges(Per Student)"])["Percentage Passing math"].mean()
spending_passing_reading = school_spending_df.groupby(["Spending Ranges(Per Student)"])["Percentage Passing reading"].mean()
overall_passing_spending = school_spending_df.groupby(["Spending Ranges(Per Student)"])["Percantage Overall Passing"].mean()
```

```python
spending_summary = pd.DataFrame({
    "Average math score": spending_math_scores,
    "Average reading score": spending_reading_scores,
    "Percentage Passing math": spending_passing_math,
    "Percentage Passing reading": spending_passing_reading,
    "Percantage Overall Passing": overall_passing_spending
})

spending_summary["Average math score"] = spending_summary["Average math score"].map("{:,.2f}".format)
spending_summary["Average reading score"] = spending_summary["Average reading score"].map("{:,.2f}".format)
spending_summary["Percentage Passing math"] = spending_summary["Percentage Passing math"].map("{:,.2f}".format)
spending_summary["Percentage Passing reading"] = spending_summary["Percentage Passing reading"].map("{:,.2f}".format)
spending_summary["Percantage Overall Passing"] = spending_summary["Percantage Overall Passing"].map("{:,.2f}".format)


spending_sum = pd.DataFrame(spending_summary)
spending_sum
```

# Scores by School Size

```python
size_bins = [0, 1000, 2000, 5000]
labels = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]
```

```python
school_spending_df["Total Student Ranges"] = pd.cut(total_student, bins=size_bins, labels=labels, right = False, include_lowest = True)
school_spending_df
```

```python
#Create a DataFrame called size_summary that breaks down school performance
#based on school size (small, medium, or large).

size_math_scores = school_spending_df.groupby(["Total Student Ranges"])["Average math score"].mean()
size_reading_scores = school_spending_df.groupby(["Total Student Ranges"])["Average reading score"].mean()
size_passing_math = school_spending_df.groupby(["Total Student Ranges"])["Percentage Passing math"].mean()
size_passing_reading = school_spending_df.groupby(["Total Student Ranges"])["Percentage Passing reading"].mean()
overall_passing_size = school_spending_df.groupby(["Total Student Ranges"])["Percantage Overall Passing"].mean()
```

```python
size_summary = pd.DataFrame({
    "Average math score": size_math_scores,
    "Average reading score": size_reading_scores,
    "Percentage Passing math": size_passing_math,
    "Percentage Passing reading": size_passing_reading,
    "Percantage Overall Passing": overall_passing_size
})

size_summary["Average math score"] = size_summary["Average math score"].map("{:,.2f}".format)
size_summary["Average reading score"] = size_summary["Average reading score"].map("{:,.2f}".format)
size_summary["Percentage Passing math"] = size_summary["Percentage Passing math"].map("{:,.2f}".format)
size_summary["Percentage Passing reading"] = size_summary["Percentage Passing reading"].map("{:,.2f}".format)
size_summary["Percantage Overall Passing"] = size_summary["Percantage Overall Passing"].map("{:,.2f}".format)


size_sum = pd.DataFrame(size_summary)
size_sum
```
